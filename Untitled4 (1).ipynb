```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
iris_df = pd.read_csv("iris.csv")   # upload iris.csv
iris_df.head()
```

```python
iris_df.hist(figsize=(10,6))
plt.suptitle("Distribution of Iris Numerical Features")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=iris_df.drop(columns=['variety']))
plt.title("Outlier Detection using Boxplot")
plt.show()
```

```python
sns.countplot(x='variety', data=iris_df)
plt.title("Count of Iris variety")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(iris_df.drop(columns=['variety']).corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap - Iris Dataset")
plt.show()
```
